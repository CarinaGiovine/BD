```python
#conexión a la base de datos
import pymysql

conn = pymysql.connect(
    host='127.0.0.1', user='root',
    passwd='', db='employees'
)
cursor = conn.cursor()  #cursor nos permite enviar consultas al servidor y obtener resultados
cursor.execute(
    "SELECT first_name, last_name FROM `employees` WHERE birth_date = '1956-02-12' " #consulto en la tabla empleados los que nacieron en esa fecha específica
)
r=cursor.fetchall()
for nombre, apellido in r:      #La función fetchall retorna una lista con el conjunto de filas retornadas por la última consulta ejecutada. 
    print("{0} {1}".format(nombre, apellido))
```

```python
#insertar en una tabla 
cursor = conn.cursor()
cursor.execute(
    " INSERT INTO `employees` (emp_no, birth_date,first_name, last_name,gender, hire_date) VALUES (0005, 1971-03-05, 'Roberto','Robles','M', 1995-12-20) "
)
conn.commit()   # con commit se guardan los valores en la tabla
cursor.execute(
    " SELECT emp_no, first_name FROM `employees` WHERE  last_name='Robles'   "   #hago una consulta para ver si se agrego el valor que agregue en el insert
)
resul=cursor.fetchall()
for i in resul:
    print(i)
```

```python
#actualizar un registro en la bd
cursor = conn.cursor()
cursor.execute(
    " UPDATE `employees` SET gender='F' WHERE emp_no=10005 "   #actualizo un dato del registro de empleado nro 10005
)
conn.commit()
cursor.execute(
    " SELECT * FROM `employees` WHERE emp_no= 10005"   # hago una consulta para verificar si se actualizo el registro
)
resu=cursor.fetchone()
resu
```

```python
#eliminar un registro de la tabla employees
cursor = conn.cursor()
cursor.execute(
    " DELETE  FROM `employees` WHERE emp_no= 5"    #elimino de la BD el registro cuyo emp_no=5
)
conn.commit()
cursor.execute(
    "SELECT * FROM `employees` WHERE emp_no= 5"   #consulto la base para ver si aun está el registro que elimine en el query de arriba
)
re=cursor.fetchone()
re
```

```python
#consulto los empleados del area de finanzas que están actualmente trabajando con INNER JOIN
cursor = conn.cursor()
cursor.execute(
    "SELECT e.first_name, e.last_name, de.from_date, de.to_date FROM dept_emp de INNER JOIN employees e ON e.emp_no = de.emp_no where de.dept_no = 'd002' AND de.to_date >= NOW()"
)
r=cursor.fetchall()
for i in r:  #recorro los rsultados y los muestro de manera mas ordenada
    print(i)
```

```python
#calculo del promedio de salarios del género Masculino, de los empleados del depto d002 que es Development, consulta que involucra varias tablas
cursor = conn.cursor()
cursor.execute(
    "SELECT ROUND(AVG(salary),0) FROM  salaries, employees, dept_emp WHERE employees.gender='M' AND employees.emp_no = salaries.emp_no AND employees.emp_no = dept_emp.emp_no AND dept_emp.dept_no = 'd002' "
)
res_m=cursor.fetchone()
res_m
```

```python
#calculo del promedio de salarios del género Femenino, de los empleados del depto d002 que es Development
cursor = conn.cursor()
cursor.execute(
    "SELECT ROUND(AVG(salary),0) FROM  salaries, employees, dept_emp WHERE employees.gender='F' AND employees.emp_no = salaries.emp_no AND employees.emp_no = dept_emp.emp_no AND dept_emp.dept_no = 'd002' "
)
res_f=cursor.fetchone()
res_f
```

```python
#promedio de salarios por departamento de los empleados actuales
cursor = conn.cursor()
cursor.execute(
    "SELECT AVG(s.salary), d.dept_name FROM salaries s, employees e, dept_emp de, departments d WHERE e.emp_no = s.emp_no AND e.emp_no = de.emp_no AND de.dept_no = d.dept_no GROUP BY d.dept_name "
)
res_pro=cursor.fetchall()
res_pro
for i in res_pro:
    print(i)
```

```python
# busco la cantidad de empleados en los que su apellido termina en 'boom'
cant=0
cursor = conn.cursor()
cursor.execute(
    "SELECT emp_no, first_name, last_name FROM `employees` WHERE last_name LIKE '%boom' "
)
b=cursor.fetchall()

for i in b:
    print(i)
    cant=cant + 1
    
print("La cantidad de empleados que sus apellidos terminan en 'boom' son: ", cant)
```
